```python
import tkinter as tk
from tkinter import filedialog
import random
import string
from PIL import Image, ImageTk
```

```python
def generate():
    length = int(entry.get())
    chars = string.ascii_letters + string.digits + "!@#$%"
    password = ''.join(random.choice(chars) for _ in range(length))
    result.config(text=password)

root = tk.Tk()
root.title("Password Generator")

tk.Label(root, text="Length").pack()
entry = tk.Entry(root)
entry.pack()

tk.Button(root, text="Generate", command=generate).pack()

result = tk.Label(root, text="", font=("Arial", 14))
result.pack()

root.mainloop()
```

```python
def open_image():
    file = filedialog.askopenfilename()
    img = Image.open(file)
    img = img.resize((300,300))
    photo = ImageTk.PhotoImage(img)
    label.config(image=photo)
    label.image = photo

root = tk.Tk()
root.title("Image Viewer")

tk.Button(root, text="Open Image", command=open_image).pack()
label = tk.Label(root)
label.pack()

root.mainloop()
```
