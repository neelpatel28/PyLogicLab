```python
import tkinter as tk
from tkinter import messagebox

class MissionariesCannibals:
    def __init__(self, root):
        self.root = root
        self.root.title("🏝️ Missionaries & Cannibals")
        self.root.geometry("900x600")
        self.root.resizable(False, False)

        self.canvas = tk.Canvas(root, width=900, height=450, bg="lightblue")
        self.canvas.pack()

        self.control_frame = tk.Frame(root)
        self.control_frame.pack(pady=10)

        self.reset_game()
        self.create_controls()
        self.draw_scene()

    def reset_game(self):
        self.left_m = 3
        self.left_c = 3
        self.right_m = 0
        self.right_c = 0
        self.boat_side = "left"

    def create_controls(self):
        tk.Label(self.control_frame, text="Missionaries:").grid(row=0, column=0)
        self.m_spin = tk.Spinbox(self.control_frame, from_=0, to=2, width=5)
        self.m_spin.grid(row=0, column=1)

        tk.Label(self.control_frame, text="Cannibals:").grid(row=0, column=2)
        self.c_spin = tk.Spinbox(self.control_frame, from_=0, to=2, width=5)
        self.c_spin.grid(row=0, column=3)

        tk.Button(self.control_frame, text="🚣 Move Boat", command=self.move_boat).grid(row=0, column=4, padx=10)
        tk.Button(self.control_frame, text="🔄 Restart", command=self.restart).grid(row=0, column=5)

    def draw_person(self, x, y, color):
        # Head
        self.canvas.create_oval(x-10, y-30, x+10, y-10, fill=color)
        # Body
        self.canvas.create_line(x, y-10, x, y+20, width=2)
        # Arms
        self.canvas.create_line(x-15, y, x+15, y, width=2)
        # Legs
        self.canvas.create_line(x, y+20, x-10, y+35, width=2)
        self.canvas.create_line(x, y+20, x+10, y+35, width=2)

    def draw_scene(self):
        self.canvas.delete("all")

        # River
        self.canvas.create_rectangle(350, 0, 550, 450, fill="blue")

        # Left Ground
        self.canvas.create_rectangle(0, 350, 350, 450, fill="green")

        # Right Ground
        self.canvas.create_rectangle(550, 350, 900, 450, fill="green")

        # Draw Left Side People
        for i in range(self.left_m):
            self.draw_person(80 + i*60, 350, "white")
        for i in range(self.left_c):
            self.draw_person(80 + i*60, 300, "red")

        # Draw Right Side People
        for i in range(self.right_m):
            self.draw_person(620 + i*60, 350, "white")
        for i in range(self.right_c):
            self.draw_person(620 + i*60, 300, "red")

        # Draw Boat
        if self.boat_side == "left":
            self.canvas.create_rectangle(280, 320, 350, 350, fill="brown")
        else:
            self.canvas.create_rectangle(550, 320, 620, 350, fill="brown")

    def is_valid(self):
        if self.left_m < 0 or self.left_c < 0 or self.right_m < 0 or self.right_c < 0:
            return False

        if self.left_m > 0 and self.left_c > self.left_m:
            return False

        if self.right_m > 0 and self.right_c > self.right_m:
            return False

        return True

    def move_boat(self):
        m = int(self.m_spin.get())
        c = int(self.c_spin.get())

        if m + c == 0 or m + c > 2:
            messagebox.showerror("Invalid Move", "Boat must carry 1 or 2 people.")
            return

        if self.boat_side == "left":
            self.left_m -= m
            self.left_c -= c
            self.right_m += m
            self.right_c += c
            self.boat_side = "right"
        else:
            self.right_m -= m
            self.right_c -= c
            self.left_m += m
            self.left_c += c
            self.boat_side = "left"

        if not self.is_valid():
            messagebox.showerror("Game Over", "Missionaries were eaten! ❌")
            self.reset_game()

        if self.right_m == 3 and self.right_c == 3:
            messagebox.showinfo("Victory!", "🎉 Everyone crossed safely!")
            self.reset_game()

        self.draw_scene()

    def restart(self):
        self.reset_game()
        self.draw_scene()

if __name__ == "__main__":
    root = tk.Tk()
    game = MissionariesCannibals(root)
    root.mainloop()
```
