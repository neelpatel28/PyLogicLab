```python
import tkinter as tk
import random
import time
```

```python
def change_color():
    colors = ["red","blue","green","yellow","purple","orange"]
    root.configure(bg=random.choice(colors))

root = tk.Tk()
root.title("Color Changer")
root.geometry("400x300")

tk.Button(root, text="Change Background", command=change_color).pack(pady=100)

root.mainloop()
```

```python
def update_time():
    current = time.strftime("%H:%M:%S")
    label.config(text=current)
    label.after(1000, update_time)

root = tk.Tk()
root.title("Digital Clock")

label = tk.Label(root, font=("Arial", 40))
label.pack()

update_time()
root.mainloop()
```
